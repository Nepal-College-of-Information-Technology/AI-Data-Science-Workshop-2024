## 2. Creating Basic Plots

### Line Plots

Line plots are useful to visualize data trends over a continuous range, such as time.

Example:

````python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

#### Understanding Markers in Matplotlib

In Matplotlib, the `marker` parameter is used to specify the style of markers that appear at each data point in a plot. For example, `marker='o'` uses a circular marker. 

Here are some common marker styles:
- **`'o'`**: Circle marker
- **`'s'`**: Square marker
- **`'^'`**: Triangle up marker
- **`'D'`**: Diamond marker
- **`'*'`**: Star marker
- **`'x'`**: X marker

Markers help to highlight individual data points, making plots more informative and visually engaging.

### Bar Charts

Bar charts are useful when you need to compare different categories.

Example:

````python
categories = ['Category A', 'Category B', 'Category C']
values = [5, 7, 3]

plt.bar(categories, values, color='skyblue')
plt.title("Simple Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
````

#### Understanding Colors in Matplotlib

In Matplotlib, the `color` parameter is used to specify the color of elements in your plots, such as lines, bars, or markers. For instance, `color='skyblue'` sets the color of the bars in a bar chart to a light blue shade.

Here are some ways to specify colors in Matplotlib:
- **Color Names**: Matplotlib supports a wide range of named colors, such as `'skyblue'`, `'red'`, `'green'`, `'orange'`, etc.
- **Hex Codes**: You can use hex color codes (e.g., `'#1f77b4'`) for precise color choices.
- **RGB/RGBA Tuples**: Colors can also be defined using RGB or RGBA tuples, like `(0.1, 0.2, 0.5)` for custom colors.

Using colors effectively in your plots can help distinguish between different data series and make your visualizations more appealing and easier to interpret.

### Scatter Plots

Scatter plots are used to visualize the relationship between two continuous variables.

Example:

````python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

### Creating a Histogram

Histograms are used to visualize the distribution of a dataset. They are useful for understanding the frequency of different ranges of values in a dataset.

````python
# Sample data: random ages
ages = [22, 25, 29, 22, 23, 30, 24, 28, 27, 22, 26, 25, 24, 29, 30]

# Creating a histogram
plt.hist(ages, bins=8, color='lightgreen', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
````

### Pie Charts

Pie charts are useful for showing the proportions of categories in a dataset. For example, showing the distribution of exam grades.

````python
labels = ['A', 'B', 'C', 'D', 'F']
sizes = [15, 30, 45, 7, 3]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0.1, 0, 0, 0, 0)  # explode the 'A' slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Grade Distribution in Exams")
plt.show()
````

## 3. Customizing Plots

### Titles, Labels, and Legends

You can add titles, axis labels, and legends to make the plots more informative.

Example:

````python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], label="Line 1")
plt.plot([1, 2, 3, 4], [40, 50, 60, 70], label="Line 2")
plt.title("Customized Plot with Legends")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
````

### Changing Colors and Styles

You can customize the appearance of your plots by changing the colors, line styles, and markers.

Example:

````python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='green', linestyle='--', marker='o')
plt.title("Styled Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

## 4. Practical Examples with Real Data

### Example 1: Visualizing Student Exam Results

Lets assume we have the final exam scores of 5 students from NCIT. Well use a bar chart to visualize their performance.

````python
students = ['Ram', 'Sita', 'Hari', 'Gita', 'Aarav']
scores = [85, 90, 75, 80, 92]

plt.bar(students, scores)
plt.title("Student Exam Results")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
````

In this example, we can see a comparison of the exam scores of five students.


### Example 2: Visualizing Population Data of Nepali Cities

Now, lets use a bar chart to visualize the population data of a few major cities in Nepal.

````python
cities = ['Kathmandu', 'Pokhara', 'Lalitpur', 'Biratnagar', 'Butwal']
population = [1000000, 500000, 300000, 250000, 200000]

plt.bar(cities, population)
plt.title("Population of Major Cities in Nepal")
plt.xlabel("Cities")
plt.ylabel("Population")
plt.show()
````

This plot visualizes the population of major Nepali cities, helping students understand the distribution of population across different regions.

### Example 3: Line Plot for Temperature Trends in Kathmandu

Lets visualize the temperature trends over a week in Kathmandu using a line plot.

````python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
temperature = [15, 16, 17, 19, 20, 18, 17]

plt.plot(days, temperature, marker='o', color='blue')
plt.title("Temperature Trends in Kathmandu Over a Week")
plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.grid(True)
plt.show()
````

### Example 4: Pie Chart for Exam Grade Distribution

This example visualizes the grade distribution in a class.

````python
labels = ['A', 'B', 'C', 'D', 'F']
sizes = [15, 30, 45, 7, 3]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0.1, 0, 0, 0, 0)  # explode the 'A' slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Grade Distribution in Exams")
plt.show()
````

### Example 5: Plotting histogram of an image data

This example visualizes histogram of the given image.

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the image (replace with your local path)
image_path = "dataset/lena.png"
image = Image.open(image_path)

# Convert the image to grayscale
image_gray = image.convert('L')

# Convert the image to a numpy array
image_array = np.array(image_gray)

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(image_array.flatten(), bins=256, range=(0, 256), color='gray', alpha=0.7)
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
````
