# Part 3: Hierarchical Clustering

In this notebook, we will explore **Hierarchical Clustering**, another popular unsupervised learning algorithm. Unlike K-Means, hierarchical clustering builds a hierarchy of clusters that can be represented in a tree structure known as a **dendrogram**.

---

## What is Hierarchical Clustering?

**Hierarchical Clustering** is a method of clustering that seeks to build a hierarchy of clusters. It comes in two flavors:

1. **Agglomerative (Bottom-up)**: 
   - Start with each data point as its own cluster and progressively merge clusters together.
   
2. **Divisive (Top-down)**:
   - Start with all data points in a single cluster and progressively split them into smaller clusters.

In this notebook, we will focus on **Agglomerative Clustering**, which is more commonly used.

---

## Key Steps in Agglomerative Clustering:

1. **Compute the distance matrix**: Calculate the distance between each pair of points.
2. **Merge closest clusters**: Identify the two closest clusters and merge them.
3. **Update the distance matrix**: Recalculate the distance between the new cluster and the remaining clusters.
4. **Repeat**: Continue merging clusters until only one cluster remains.
5. **Dendrogram**: Visualize the merging process using a dendrogram.

---

## Step 1: Load the Iris Dataset

We will use the **Iris dataset** for hierarchical clustering as well. Let’s load and visualize the dataset.

```python
# Import necessary libraries
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Display the first few rows of the dataset
df.head()
```

## Step 2: Perform Hierarchical Clustering

We will now perform **agglomerative hierarchical clustering** using Scikit-learn's `AgglomerativeClustering` class.

### Linkage Criteria

- **Single Linkage**: Minimum distance between clusters.
- **Complete Linkage**: Maximum distance between clusters.
- **Average Linkage**: Average distance between clusters.

We will apply **complete linkage** for this clustering example.

```python
from sklearn.cluster import AgglomerativeClustering

# Perform hierarchical clustering with complete linkage
hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='complete')
df['Cluster'] = hc.fit_predict(df)

# Display the first 80 rows with cluster labels
df.head(80)
```

## Step 3: Visualizing the Dendrogram

The **dendrogram** is a tree-like diagram that shows the merging process of the hierarchical clustering algorithm. It is useful for visualizing the structure of the clusters and determining the optimal number of clusters.

```python
# Import necessary libraries for dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage

# Perform linkage
Z = linkage(df.iloc[:, :-1], method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

## Step 4: Visualizing the Clusters

Now, let’s visualize the clusters formed by the hierarchical clustering algorithm using a scatter plot of the first two features (sepal length and sepal width).

```python
# Plot the clusters
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['Cluster'], cmap='rainbow', s=50)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hierarchical Clustering of Iris Data (Complete Linkage)')
plt.show()
```

### Conclusion

In this notebook, we:
- Introduced the **Hierarchical Clustering** algorithm.
- Performed **Agglomerative Clustering** on the **Iris dataset**.
- Visualized the clusters using a **dendrogram** and a scatter plot.

Hierarchical clustering is a useful method when you need to create a hierarchy of clusters and visualize the merging process. The dendrogram can help decide the optimal number of clusters based on the merging distances.
